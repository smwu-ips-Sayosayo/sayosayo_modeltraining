```python
import os

# 레이블 파일이 저장된 폴더 지정
folder_path = '/content/drive/MyDrive/Colab Notebooks/data/dataset5_processed2/label'

# 고유한 내용을 저장할 집합
unique_contents = set()

# 폴더 내 모든 .txt 파일을 순회
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):  # 레이블 파일 확인
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            lines = file.readlines()
            for line in lines:
                # 전체 라인을 고유한 내용으로 간주하고 저장
                unique_contents.add(line.strip())

# 고유한 내용 출력
print(f'Found {len(unique_contents)} unique lines in label files:')
for content in unique_contents:
    print(content)
```

```python
import os
import random
import shutil

# 원본 데이터 폴더 경로 (이미지와 레이블이 같은 폴더에 있는 경우)
data_folder = "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/label"

# 경로 검증 및 폴더 생성
folders = [
    "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/train/labels",
    "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/train/images",
    "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/test/labels",
    "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/test/images"
]

for folder in folders:
    if not os.path.exists(folder):
        os.makedirs(folder)

# 훈련 및 검증 데이터 분할
files = [f for f in os.listdir(data_folder) if f.endswith('.txt')]  # 레이블 파일 기준으로 리스트 생성
random.shuffle(files)

train_size = int(len(files) * 0.8)
train_files = files[:train_size]
validation_files = files[train_size:]

# 파일 이동 로직
def move_files(file_list, source_folder, dest_image_folder, dest_label_folder):
    for file_name in file_list:
        label_path = os.path.join(source_folder, file_name)
        image_path = label_path.replace('.txt', '.jpg')  # JPG 파일 가정

        if os.path.exists(label_path) and os.path.exists(image_path):
            shutil.move(label_path, dest_label_folder)
            shutil.move(image_path, dest_image_folder)
        else:
            print(f"Missing file: {label_path} or {image_path}")

# 훈련 데이터 이동
move_files(train_files, data_folder,
           "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/train/images",
           "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/train/labels")

# 검증 데이터 이동
move_files(validation_files, data_folder,
           "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/test/images",
           "/content/drive/MyDrive/Colab Notebooks/data/dataset1_snack/test/labels")

print("데이터 분할이 완료되었습니다.")
```
